```python
import seaborn as sns
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

```python
# Load the dataset

titanic = sns.load_dataset("titanic")
fmri = sns.load_dataset("fmri")
tips = sns.load_dataset("tips")
flights = sns.load_dataset("flights")
```

### Scatter Plots

```python
sns.scatterplot(data=titanic, x="age", y="fare")
```

#### Specifiy the categories distinction using hue feature

```python
sns.scatterplot(data=titanic, x="age", y="fare", hue="sex")
```

```python
sns.scatterplot(data=titanic, x="age", y="fare", hue="who")
```

```python
sns.scatterplot(data=titanic, x="age", y="fare", hue="age")
```

```python
sns.scatterplot(data=titanic, x="age", y="fare", hue="sex", style="sex")
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=titanic, x="age", y="fare", hue="sex", style="sex")
```

```python
plt.figure(figsize=(8,6))



sns.scatterplot(data=titanic, x="age", y="fare", hue="sex", size="pclass", sizes=(20, 200), style="sex")
```
