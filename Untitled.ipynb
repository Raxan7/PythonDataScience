```python
# Concatenation

data_a = {
    "id": ['1', '2', '3'],
    "first_name": ["Alex", "Amy", "Allen"]
}
df_a = pd.DataFrame(data=data_a, columns=["id", "first_name"])
df_a
```

```python
# Concatenation

data_b = {
    "id": ['1', '2', '3'],
    "last_name": ["Lucha", "Amy", "Allen"]
}
df_b = pd.DataFrame(data=data_b, columns=["id", "last_name"])
df_b
```

```python
pd.merge(df_a, df_b, how="outer", on="id")
```

```python
pd.merge(df_a, df_b, how="inner", on="id")
```

```python
pd.concat([df_a, df_b], axis=0)
```
